```python
import numpy as np

matrix = np.random.randint(1, 11, size=(3, 3))

print("Случайная матрица 3×3:")
print(matrix)

det = np.linalg.det(matrix)

if abs(det) != 0:  
    inv_matrix = np.linalg.inv(matrix)
    print("\nОбратная матрица существует:")
    print(inv_matrix)

    prover = matrix @ inv_matrix
    print("\nПроверка: A × A⁻¹ = E (единичная матрица):")
    print(np.round(prover, 2))
else:
    print("\nМатрица вырождена (определитель = 0), обратная матрица не существует!")
```
