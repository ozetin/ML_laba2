```python
import numpy as np

matrix1 = np.random.randint(1, 21, size=(3, 3))
matrix2 = np.random.randint(1, 21, size=(3, 3))

print(matrix1)
print(matrix2)

sum_matrix = matrix1 + matrix2
print("\nСумма матриц (A + B):")
print(sum_matrix)

razn_matrix = matrix1 - matrix2
print("\nРазность матриц (A - B):")
print(razn_matrix)

# Вычисляем поэлементное произведение
proizv = matrix1 * matrix2
print("\nПоэлементное произведение (A * B):")
print(proizv)
```
