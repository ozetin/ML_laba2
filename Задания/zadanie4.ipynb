```python
import numpy as np

arr = np.linspace(-10, 10, 50)

print("Массив из 50 чисел от -10 до 10:")
print(arr)

summa = np.sum(arr)
print(f"Сумма всех элементов: {summa}")

polozh_summa = np.sum(arr[arr > 0])
print(f"Сумма положительных элементов: {polozh_summa}")

otriz_sum = np.sum(arr[arr < 0])
print(f"Сумма отрицательных элементов: {otriz_sum}")
```
