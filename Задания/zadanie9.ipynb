```python
import numpy as np

matrix = np.random.randint(1, 51, size=(4, 4))

print("Исходная матрица (4×4):")
print(matrix)

transpon_matrix = matrix.T  
print("\nТранспонированная матрица:")
print(transpon_matrix)

trace = np.trace(matrix)
print(f"\nСлед матрицы (сумма элементов на главной диагонали): {trace}")
```
