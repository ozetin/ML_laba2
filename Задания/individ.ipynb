```python
import numpy as np

A = np.array([[-1.5, 1, 0],
              [0, -1.2, 1],
              [1, 1, 1]])

B = np.array([0, 0, 5_000_000])

matr = np.linalg.inv(A) @ B

det_A = np.linalg.det(A)

def kram(A, B, col):
    A_i = A.copy()
    A_i[:, col] = B
    return np.linalg.det(A_i) / det_A

kramer = np.array([kram(A, B, i) for i in range(3)])

solve = np.linalg.solve(A, B)

print("Матричный метод:", matr)
print("Метод Крамера:  ", kramer)
print("np.linalg.solve: ", solve)
```
